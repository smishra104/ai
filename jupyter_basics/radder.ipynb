```python
import os
import openai
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv(), override=True)
openai.api_key = os.getenv('OPEN_API_KEY')
```

```python
from openai import OpenAI
client = OpenAI()

def get_tts(prompt):
    response = client.audio.speech.create(
        model='tts-1',
        voice='alloy',
        input=prompt
    )
    return response
    
```

```python
response = get_tts('Didi is moti, more than moti, motier than moti')
print(response)
response.stream_to_file('dd.mp3')
import IPython
IPython.display.Audio('dd.mp3')
```
